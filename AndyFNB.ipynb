```
"""
Code to calculate when a paper is folded 43 times
"""
THICKNESS = 0.00008
folded_thickness = THICKNESS*43
print("Thickness: {} meters".format(folded_thickness))
```

```
THICKNESS = 0.00008
n_folded = 43
folded_thickness = THICKNESS
for i in range(n_folded):
  folded_thickness = folded_thickness*43
  print("Thickness: {} meters".format(folded_thickness))
```

```
print("Thickness: {: .2f} km".format(folded_thickness/1000))
```

```
%%timeit
import time
THICKNESS = 0.00008
n_folded = 43
start = time.time()
folded_thickness = THICKNESS*(43**n_folded)
elapsed_time = time.time() - start
print("time: {}[ms]".format(elapsed_time*1000))
```

```
%%%timeit
import time
THICKNESS = 0.00008
n_folded = 43
start = time.time()
folded_thickness = THICKNESS
for i in range(n_folded):
  folded_thickness = folded_thickness*43
  elapsed_time = time.time() - start
  print("time: {}[ms]".format(elapsed_time*1000))
```

```
'''
The relationship between the two calculated time, is that any time you run the code
the time changes, but still remain within the same range of the previous calculations 
with a small difference. 
The for statement of writting seems a better way of writting and it is faster to 
calculate.
'''
```

```
THICKNESS = 0.00008
n_folded = 43
folded_thickness_list = [THICKNESS]
for i in range(n_folded):
  folded_thickness_list.append(folded_thickness_list[-1]*43)
  print("Thickness: {}meters".format(folded_thickness_list[-1]))
  print("--------")
  print("list iteam")
  print(folded_thickness_list)
  
 
  
```

```
import matplotlib.pyplot as plt
%matplotlib inline
plt.title("Thickness of a Folded Paper")
plt.xlabel("Number of Folds")
plt.ylabel("Thickness[m]")
plt.plot(folded_thickness_list) 
plt.show()
```

```
'''
At about 40 folds the thickness start rising up to 0.1, and by 42 to 43 it 
start moving up from 0.1 to 0.3.... etc. As the number of fold increases so 
is the thickness increasing.
'''
```

```
'''
Display a Red Line Graph.
'''
plt.title("thickness of folded paper")
plt.xlabel("number of folds")
plt.ylabel("thickness[m]")
plt.plot(folded_thickness_list, color='red')
plt.show()
```

```
plt.title("Thickness of Folded Paper")
plt.xlabel("Number of Folds")
plt.ylabel("Thickness[m]")
plt.tick_params(labelsize=20)
plt.plot(folded_thickness_list, color='green')
plt.show()
```
